```python
import numpy as np
import matplotlib.pyplot as plt

# Extract data
X3 = df[["Kilometer"]].values
y3 = df["Price"].values.reshape(-1,1)

# Add bias column (intercept term)
X3_b = np.c_[np.ones((len(X3), 1)), X3]  # shape (n,2)

# Gradient Descent
n_iter = 1000
alpha = 0.000000000001   # ⚠ adjust learning rate (Kilometer values are large!)
m = len(y3)

theta = np.random.randn(2, 1)  # [theta0, theta1]
costs = []

for iteration in range(n_iter):
    gradients = (2/m) * X3_b.T.dot(X3_b.dot(theta) - y3)
    theta = theta - alpha * gradients
    cost = (1/(2*m)) * np.sum((X3_b.dot(theta) - y3) ** 2)
    costs.append(cost)

print("Theta found by gradient descent:", theta.ravel())

# Closed form solution (Normal Equation)
theta_best = np.linalg.inv(X3_b.T.dot(X3_b)).dot(X3_b.T).dot(y3)
print("Theta (normal eq):", theta_best.ravel())

# Plot regression line
plt.scatter(X3, y3, alpha=0.6, label="data")
plt.plot(X3, X3_b.dot(theta), color="red", label="gradient descent line")
plt.plot(X3, X3_b.dot(theta_best), color="green", linestyle="--", label="normal eq line")
plt.legend()
plt.xlabel("Kilometer")
plt.ylabel("Price")
plt.show()

# Plot cost function decrease
plt.plot(range(n_iter), costs)
plt.xlabel("Iteration")
plt.ylabel("Cost (MSE)")
plt.title("Gradient Descent Convergence")
plt.show()
```
